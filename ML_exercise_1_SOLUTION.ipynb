# A single layer ANN exercise

In this python notebook you should:
1. Import a dataset from a text file
2. Split the dataset into input and ouput data
3. Build a single layer ANN using tensorflow and keras
4. Visualize the accuracy of the ANN
5. What can you try to improve the accuracy of the ANN?

#### Dataset description

The dataset we are interested in is stored in the file called 'Data_Example_1.txt'.\
The file contains 1000 points where the three first columns represent the input variable while the 3 last columns represents the ouput variables.

#### ANN description

The ANN we want to build contains 3 input and 3 output variables.\
The activation functions should be defined as rectified linear units (relu).\
The output layer should be defined by a linear activation function.\
As a starting point the number of neurons should be set to 50.\
The number of epochs should be set to 50.

---

```python
# We start by importing the required modules
import numpy as np
# import matplotlib.pyplot as plt
import tensorflow as tf
```

We read the data from the text file called 'Data_Exercise_1.txt' using numpy

```python
data = np.loadtxt('Data_Exercise_1.txt')
```

We can now split the input and output data.

```python
input_data, output_data = data[:,0:3],data[:,3:6]
```

Next, we build a first ANN using 50 neurons and the relu activation function

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(units=50, activation='relu', input_dim=3))
model.add(tf.keras.layers.Dense(units=3, activation='linear'))
model.compile(loss='mse', optimizer='adam')
```

The training of the ANN is first carried out using 50 epochs.

```python
history = model.fit(input_data, output_data, epochs=50, verbose=0)
```

To evaluate the results from the network we plot the predicted ouputs compared to original data.

```python
plt.rcParams["figure.figsize"] = (30, 10)
fig,axs = plt.subplots(1,3)
#
prediction = model.predict(input_data)
#
axs[0].scatter(output_data[:,0],prediction[:,0])
axs[1].scatter(output_data[:,1],prediction[:,1])
axs[2].scatter(output_data[:,2],prediction[:,2])
#
axs[0].plot([0.0,np.max(output_data[:,0])],[0.0,np.max(prediction[:,0])],'r')
axs[1].plot([0.0,np.max(output_data[:,1])],[0.0,np.max(prediction[:,1])],'r')
axs[2].plot([0.0,np.max(output_data[:,2])],[0.0,np.max(prediction[:,2])],'r')
#
axs[0].set_xlabel('Measured')
axs[0].set_ylabel('Predicted')
axs[0].grid()

axs[1].set_xlabel('Measured')
axs[1].set_ylabel('Predicted')
axs[1].grid()

axs[2].set_xlabel('Measured')
axs[2].set_ylabel('Predicted')
axs[2].grid()
```

We see that the network is giving decent predictions but is far from being perfectly able to predict the data.
The first attempt to improve the prediction can be to increase the number of epochs used for training.
This can be done by simply executing a new cell as follow:

```python
history = model.fit(input_data, output_data, epochs=150, verbose=0)
```

```python
plt.rcParams["figure.figsize"] = (30, 10)
fig,axs = plt.subplots(1,3)
#
prediction = model.predict(input_data)
#
axs[0].scatter(output_data[:,0],prediction[:,0])
axs[1].scatter(output_data[:,1],prediction[:,1])
axs[2].scatter(output_data[:,2],prediction[:,2])
#
axs[0].plot([0.0,np.max(output_data[:,0])],[0.0,np.max(prediction[:,0])],'r')
axs[1].plot([0.0,np.max(output_data[:,1])],[0.0,np.max(prediction[:,1])],'r')
axs[2].plot([0.0,np.max(output_data[:,2])],[0.0,np.max(prediction[:,2])],'r')
#
axs[0].set_xlabel('Measured')
axs[0].set_ylabel('Predicted')
axs[0].grid()

axs[1].set_xlabel('Measured')
axs[1].set_ylabel('Predicted')
axs[1].grid()

axs[2].set_xlabel('Measured')
axs[2].set_ylabel('Predicted')
axs[2].grid()
```

In this example we see that increasing the number of epochs from 50 to 200 gives a significant improvement of the fit.
